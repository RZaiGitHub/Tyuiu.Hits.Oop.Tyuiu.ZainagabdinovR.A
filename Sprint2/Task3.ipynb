Создать базовый класс Product в C#, который будет представлять информацию о 
продуктах. На основе этого класса разработать 2-3 производных класса, 
демонстрирующих принципы наследования и полиморфизма. В каждом из классов 
должны быть реализованы новые атрибуты и методы, а также переопределены 
некоторые методы базового класса для демонстрации полиморфизма. 

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

<h2 style="color:DodgerBlue">Реализация:</h2>

```
using System;

// Базовый класс Employee
public class Employee
{
    public string Name { get; set; }
    public int Age { get; set; }
    public double Salary { get; set; }
    public string Department { get; set; } // Добавленный атрибут
    public DateTime HireDate { get; set; } // Добавленный атрибут
    public string Address { get; set; } // Добавленный атрибут
    public string Email { get; set; } // Добавленный атрибут


    public Employee(string name, int age, double salary, string department, DateTime hireDate, string address, string email)
    {
        Name = name;
        Age = age;
        Salary = salary;
        Department = department;
        HireDate = hireDate;
        Address = address;
        Email = email;
    }

    public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, Зарплата: {Salary:C}, Отдел: {Department}, Дата найма: {HireDate:d}, Адрес: {Address}, Email: {Email}";
    }

    public virtual void Work()
    {
        Console.WriteLine($"{Name} выполняет свои рабочие обязанности.");
    }

    public virtual void TakeVacation()
    {
        Console.WriteLine($"{Name} берет отпуск.");
    }

    public double CalculateBonus(double percentage) // Добавленный метод
    {
        return Salary * percentage;
    }

    public void PrintAddress() // Добавленный метод
    {
        Console.WriteLine($"Адрес {Name}: {Address}");
    }
}

// Производный класс Manager (простое наследование)
public class Manager : Employee
{
    public int SubordinatesCount { get; set; }
    public string ManagementStyle { get; set; } // Добавленный атрибут


    public Manager(string name, int age, double salary, string department, DateTime hireDate, string address, string email, int subordinatesCount, string managementStyle)
        : base(name, age, salary, department, hireDate, address, email)
    {
        SubordinatesCount = subordinatesCount;
        ManagementStyle = managementStyle;
    }

    public override void Work()
    {
        Console.WriteLine($"{Name} управляет командой из {SubordinatesCount} подчиненных, используя стиль {ManagementStyle}.");
    }

    public void HoldMeeting() // Добавленный метод
    {
        Console.WriteLine($"{Name} проводит совещание.");
    }
}

// Производный класс Developer (простое наследование)
public class Developer : Employee
{
    public string Specialization { get; set; }
    public string ProgrammingLanguage { get; set; } // Добавленный атрибут


    public Developer(string name, int age, double salary, string department, DateTime hireDate, string address, string email, string specialization, string programmingLanguage)
        : base(name, age, salary, department, hireDate, address, email)
    {
        Specialization = specialization;
        ProgrammingLanguage = programmingLanguage;
    }

    public override void TakeVacation()
    {
        Console.WriteLine($"{Name} берет отпуск, согласовав его с проектными сроками по {Specialization}.");
    }

    public void WriteCode() // Добавленный метод
    {
        Console.WriteLine($"{Name} пишет код на {ProgrammingLanguage}.");
    }
}

// Производный класс Designer (простое наследование)
public class Designer : Employee
{
    public string DesignStyle { get; set; }
    public string DesignSoftware { get; set; } // Добавленный атрибут


    public Designer(string name, int age, double salary, string department, DateTime hireDate, string address, string email, string designStyle, string designSoftware)
        : base(name, age, salary, department, hireDate, address, email)
    {
        DesignStyle = designStyle;
        DesignSoftware = designSoftware;
    }

    public override void Work()
    {
        Console.WriteLine($"{Name} работает над дизайн-проектами в стиле {DesignStyle}, используя {DesignSoftware}.");
    }

    public void CreateMockup() // Добавленный метод
    {
        Console.WriteLine($"{Name} создает макет.");
    }
}

// Интерфейс для сложного и множественного наследования
public interface IProjectManager
{
    void ManageProject();
}

// Пример множественного наследования (через интерфейс)
public class ProjectManager : Manager, IProjectManager
{
    public ProjectManager(string name, int age, double salary, string department, DateTime hireDate, string address, string email, int subordinatesCount, string managementStyle)
        : base(name, age, salary, department, hireDate, address, email, subordinatesCount, managementStyle) { }

    public void ManageProject()
    {
        Console.WriteLine($"{Name} управляет проектом.");
    }
}

public class Example
{
    public static void Main(string[] args)
    {
        Employee emp = new Employee("Иван Иванов", 30, 50000, "IT", new DateTime(2022, 1, 15), "ул. Ленина 1", "ivan.ivanov@example.com");
        Manager manager = new Manager("Петр Петров", 40, 80000, "Sales", new DateTime(2020, 5, 20), "ул. Пушкина 10", "petr.petrov@example.com", 5, "Демократичный");
        Developer developer = new Developer("Сидор Сидоров", 25, 60000, "IT", new DateTime(2023, 10, 1), "ул. Некрасова 15", "sidr.sidorov@example.com", "Backend", "C#");
        Designer designer = new Designer("Мария Иванова", 35, 70000, "Design", new DateTime(2021, 3, 10), "ул. Лермонтова 20", "maria.ivanova@example.com", "Модерн", "Figma");
        ProjectManager projectManager = new ProjectManager("Анна Петрова", 45, 90000, "Project Management", new DateTime(2018, 11, 5), "ул. Толстого 25", "anna.petrova@example.com", 10, "Агильный");


        Console.WriteLine(emp.GetInfo());
        emp.Work();
        emp.TakeVacation();
        Console.WriteLine($"Бонус для {emp.Name}: {emp.CalculateBonus(0.1):C}");
        emp.PrintAddress();
        Console.WriteLine();


        Console.WriteLine(manager.GetInfo());
        manager.Work();
        manager.TakeVacation();
        manager.HoldMeeting();
        Console.WriteLine();

        Console.WriteLine(developer.GetInfo());
        developer.Work();
        developer.TakeVacation();
        developer.WriteCode();
        Console.WriteLine();

        Console.WriteLine(designer.GetInfo());
        designer.Work();
        designer.TakeVacation();
        designer.CreateMockup();
        Console.WriteLine();

        Console.WriteLine(projectManager.GetInfo());
        projectManager.Work();
        projectManager.TakeVacation();
        projectManager.ManageProject();
        Console.WriteLine();
    }
}
```
