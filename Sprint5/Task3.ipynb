```
//1. Базовый класс Employee
using System;
using System.Collections.Generic;

public class Employee
{
    public string Name { get; set; }
    public int Age { get; set; }
    public decimal Salary { get; set; }
    public string Department { get; set; }
    public DateTime HireDate { get; set; }

     public delegate void WorkCompletedEventHandler(object sender, EventArgs e);

    public event WorkCompletedEventHandler WorkCompleted;

    public Employee(string name, int age, decimal salary, string department, DateTime hireDate)
    {
        Name = name;
        Age = age;
        Salary = salary;
        Department = department;
        HireDate = hireDate;
    }

   public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, Зарплата: {Salary}, Отдел: {Department}, Дата найма: {HireDate.ToShortDateString()}";
    }

    public virtual void Work()
    {
        Console.WriteLine($"{Name} выполняет стандартные рабочие обязанности.");
          OnWorkCompleted();
    }

    public virtual void TakeVacation()
    {
        Console.WriteLine($"{Name} уходит в отпуск.");
    }
    protected virtual void OnWorkCompleted()
    {
        WorkCompleted?.Invoke(this, EventArgs.Empty);
    }
     public virtual void AttendMeeting()
    {
        Console.WriteLine($"{Name} присутствует на общем собрании.");
    }
     public void ApplyBonus(decimal bonusPercentage)
    {
        Salary += Salary * bonusPercentage / 100;
        Console.WriteLine($"{Name} получил {bonusPercentage}%бонус. Новая зарплата: {Salary:C}");
    }
}
//2. Производный класс Manager

public class Manager : Employee
{
    public int SubordinatesCount { get; set; }
    public string ManagementStyle { get; set; }

    public Manager(string name, int age, decimal salary, string department, DateTime hireDate, int subordinatesCount, string managementStyle)
        : base(name, age, salary, department, hireDate)
    {
        SubordinatesCount = subordinatesCount;
         ManagementStyle = managementStyle;
    }

    public override void Work()
    {
        Console.WriteLine($"{Name} руководит командой из {SubordinatesCount} люди.");
         OnWorkCompleted();
    }

     public override string GetInfo()
        {
            return $"{base.GetInfo()}, Подчиненные считаются: {SubordinatesCount}, Стиль управления: {ManagementStyle}";
        }
    public void ConductPerformanceReview()
    {
        Console.WriteLine($"{Name} проводит анализ эффективности работы.");
    }
    public override void AttendMeeting()
    {
        Console.WriteLine($"{Name} присутствует на совещании руководства.");
    }
}
//3. Производный класс Developer

public class Developer : Employee
{
    public string Specialization { get; set; }
    public string ProgrammingLanguage { get; set; }

    public Developer(string name, int age, decimal salary, string department, DateTime hireDate, string specialization, string programmingLanguage)
        : base(name, age, salary, department, hireDate)
    {
        Specialization = specialization;
        ProgrammingLanguage = programmingLanguage;
    }

    public override void TakeVacation()
    {
        Console.WriteLine($"{Name} необходимо согласовывать отпуск со сроками реализации проекта в связи с их {Specialization} специализация.");
    }
    public override string GetInfo()
        {
            return $"{base.GetInfo()},Специализация: {Specialization}, Язык программирования: {ProgrammingLanguage}";
        }
    public void WriteCode()
    {
        Console.WriteLine($"{Name} заключается в написании кода в {ProgrammingLanguage}.");
    }
      public override void AttendMeeting()
    {
        Console.WriteLine($"{Name} участвует в техническом совещании.");
    }
}
//4. Производный класс Designer

public class Designer : Employee
{
    public string DesignStyle { get; set; }
     public string  DesignSoftware { get; set; }
    public Designer(string name, int age, decimal salary, string department, DateTime hireDate, string designStyle, string designSoftware)
        : base(name, age, salary, department, hireDate)
    {
        DesignStyle = designStyle;
        DesignSoftware = designSoftware;
    }

    public override void Work()
    {
        Console.WriteLine($"{Name} работает над дизайн-проектами с {DesignStyle} стиль.");
          OnWorkCompleted();
    }

      public override string GetInfo()
        {
            return $"{base.GetInfo()},Стиль дизайна: {DesignStyle}, Программное обеспечение для проектирования: {DesignSoftware}";
        }
    public void CreateMockups()
    {
        Console.WriteLine($"{Name} занимается созданием дизайнерских макетов.");
    }
     public override void AttendMeeting()
    {
        Console.WriteLine($"{Name} посещает совещание по дизайну.");
    }
}
//5. Использование коллекций, делегатов и событий в Program.cs

class Program
{
    static void Main(string[] args)
    {
        // Создание списка сотрудников
        List<Employee> employees = new List<Employee>();

        // Создание объектов производных классов
        Manager manager = new Manager("Джон Смит", 45, 120000, "Менеджмент", new DateTime(2010, 5, 15), 10, "Автократический");
        Developer developer = new Developer("Джейн Доу", 30, 90000, "Разработка",new DateTime(2022, 1, 20) , "Frontend", "JavaScript");
        Designer designer = new Designer("Элис Браун", 35, 85000, "Дизайн", new DateTime(2018, 7, 1), "Минимализм", "Figma");


        // Добавление сотрудников в список
        employees.Add(manager);
        employees.Add(developer);
        employees.Add(designer);

         foreach (var employee in employees)
        {
            Console.WriteLine(employee.GetInfo());
        }
        // Подписываемся на событие WorkCompleted
        foreach (var employee in employees)
        {
          employee.WorkCompleted += Employee_WorkCompleted;
        }
        // Вызов методов Work для демонстрации полиморфизма и срабатывания события
        foreach (var employee in employees)
        {
          employee.Work();
        }

        Console.WriteLine("\n--- Каникулы ---");
        // Демонстрация полиморфизма при вызове метода TakeVacation
         foreach (var employee in employees)
        {
           employee.TakeVacation();
        }
          Console.WriteLine("\n--- Бонусная заявка ---");

        // Используем делегат для начисления бонуса всем сотрудникам
        Action<decimal> bonusAction = percentage =>
        {
           foreach (var employee in employees)
           {
             employee.ApplyBonus(percentage);
           }
        };

        bonusAction(10); // Начисляем бонус в 10%

        Console.WriteLine("\n--- Встречи ---");
       
        foreach (var employee in employees)
        {
           employee.AttendMeeting();
        }
       
        Console.ReadKey();
    }

     private static void Employee_WorkCompleted(object sender, EventArgs e)
    {
        Console.WriteLine($"Работа, выполненная {((Employee)sender).Name} on {DateTime.Now}");
    }
}

    {
        // Создание списка сотрудников
        List<Employee> employees = new List<Employee>();

        // Создание объектов производных классов
        Manager manager = new Manager("Джон Смит", 45, 120000, "Менеджмент", new DateTime(2010, 5, 15), 10, "Автократический");
        Developer developer = new Developer("Джейн Доу", 30, 90000, "Разработка",new DateTime(2022, 1, 20) , "Frontend", "JavaScript");
        Designer designer = new Designer("Элис Браун", 35, 85000, "Дизайн", new DateTime(2018, 7, 1), "Минимализм", "Figma");


        // Добавление сотрудников в список
        employees.Add(manager);
        employees.Add(developer);
        employees.Add(designer);

         foreach (var employee in employees)
        {
            Console.WriteLine(employee.GetInfo());
        }
        // Подписываемся на событие WorkCompleted
        foreach (var employee in employees)
        {
          employee.WorkCompleted += Employee_WorkCompleted;
        }
        // Вызов методов Work для демонстрации полиморфизма и срабатывания события
        foreach (var employee in employees)
        {
          employee.Work();
        }

        Console.WriteLine("\n--- Каникулы ---");
        // Демонстрация полиморфизма при вызове метода TakeVacation
         foreach (var employee in employees)
        {
           employee.TakeVacation();
        }
          Console.WriteLine("\n--- Бонусная заявка ---");

        // Используем делегат для начисления бонуса всем сотрудникам
        Action<decimal> bonusAction = percentage =>
        {
           foreach (var employee in employees)
           {
             employee.ApplyBonus(percentage);
           }
        };

        bonusAction(10); // Начисляем бонус в 10%

        Console.WriteLine("\n--- Встречи ---");
       
        foreach (var employee in employees)
        {
           employee.AttendMeeting();
        }
       
        Console.ReadKey();
    }

     private static void Employee_WorkCompleted(object sender, EventArgs e)
    {
        Console.WriteLine($"Работа, выполненная {((Employee)sender).Name} on {DateTime.Now}");
    }


```
