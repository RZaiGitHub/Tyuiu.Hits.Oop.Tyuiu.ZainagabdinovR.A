```csharp
using System;
using System.Collections.Generic;

// Класс Product (Товар)
public class Product
{
    private string _name;
    private decimal _price;
    private int _quantity;
    protected string Category;
    internal string Description;

    public Product(string name, decimal price, int quantity, string category, string description)
    {
        _name = name;
        _price = price;
        _quantity = quantity;
        Category = category;
        Description = description;
    }

    public decimal GetTotalPrice()
    {
        return _price * _quantity;
    }

    public void UpdateQuantity(int amount)
    {
        _quantity += amount;
    }

    public string GetProductInfo()
    {
        return $"Название: {_name}, Цена: {_price}, Количество: {_quantity}, Категория: {Category}, Описание: {Description}";
    }
}

// Класс Seller (Продавец)
public class Seller
{
    private string _name;
    private string _employeeId;
    protected decimal Salary;
    internal string ContactInfo;
    private List<Product> _products = new List<Product>();
    public Seller(string name, string employeeId, decimal salary, string contactInfo)
    {
        _name = name;
        _employeeId = employeeId;
        Salary = salary;
        ContactInfo = contactInfo;
    }
     public void AddProduct(Product product)
    {
        _products.Add(product);
    }
    public void SellProduct(Product product, int quantity)
    {
         if (_products.Contains(product))
         {
                product.UpdateQuantity(-quantity);
          }
        else
         {
           Console.WriteLine($"Продукта {product.GetProductInfo()} нет в списке продавца");
         }
    }
    public string GetSellerInfo()
    {
        return $"Имя: {_name}, ID сотрудника: {_employeeId}, Зарплата: {Salary}, Контактная информация: {ContactInfo}";
    }
    public List<Product> GetSellerProducts()
    {
        return _products;
    }
}

// Класс Store (Магазин)
public class Store
{
    private string _storeName;
    private string _location;
    public string StoreHours;
    private List<Seller> _sellers = new List<Seller>();
     public Store(string storeName, string location, string storeHours)
    {
        _storeName = storeName;
        _location = location;
        StoreHours = storeHours;
    }
    public void AddSeller(Seller seller)
    {
        _sellers.Add(seller);
    }

    public void ListProducts()
    {
         Console.WriteLine($"Товары в магазине {_storeName}:");
         foreach (var seller in _sellers)
          {
              foreach(var product in seller.GetSellerProducts())
              {
                Console.WriteLine(product.GetProductInfo());
              }
         }
    }
     public string GetStoreInfo()
    {
       return $"Название магазина: {_storeName}, Расположение: {_location}, Часы работы: {StoreHours}";
    }
}

public class Program
{
    public static void Main(string[] args)
    {
        // Создание экземпляров классов
        Product product1 = new Product("Телефон", 500.00m, 10, "Электроника", "Смартфон последней модели");
        Product product2 = new Product("Ноутбук", 1200.00m, 5, "Электроника", "Мощный игровой ноутбук");
        Seller seller1 = new Seller("Иван Иванов", "12345", 3000.00m, "ivan@example.com");
        Store store1 = new Store("Техномир", "ул. Центральная, 1", "10:00 - 20:00");

        // Добавляем товары и продавцов
        seller1.AddProduct(product1);
        seller1.AddProduct(product2);
        store1.AddSeller(seller1);
         // Продажа товара
         seller1.SellProduct(product1, 2);
        // Вывод информации
        Console.WriteLine(product1.GetProductInfo());
        Console.WriteLine(seller1.GetSellerInfo());
        Console.WriteLine(store1.GetStoreInfo());
        store1.ListProducts();
       // попытка продажи товара, которого нет у продавца
        Product product3 = new Product("Клавиатура", 100.00m, 20, "Компьютерные аксесуары", "Эргономичная клавиатура");
        seller1.SellProduct(product3, 3);
        Console.ReadKey();
    }
}
Product product1 = new Product("Телефон", 500.00m, 10, "Электроника", "Смартфон последней модели");
        Product product2 = new Product("Ноутбук", 1200.00m, 5, "Электроника", "Мощный игровой ноутбук");
        Seller seller1 = new Seller("Иван Иванов", "12345", 3000.00m, "ivan@example.com");
        Store store1 = new Store("Техномир", "ул. Центральная, 1", "10:00 - 20:00");

        // Добавляем товары и продавцов
        seller1.AddProduct(product1);
        seller1.AddProduct(product2);
        store1.AddSeller(seller1);
         // Продажа товара
         seller1.SellProduct(product1, 2);
        // Вывод информации
        Console.WriteLine(product1.GetProductInfo());
        Console.WriteLine(seller1.GetSellerInfo());
        Console.WriteLine(store1.GetStoreInfo());
        store1.ListProducts();
       // попытка продажи товара, которого нет у продавца
        Product product3 = new Product("Клавиатура", 100.00m, 20, "Компьютерные аксесуары", "Эргономичная клавиатура");
        seller1.SellProduct(product3, 3);
        Console.ReadKey();
```
